1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
# Importando pacotes
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````python
# Importando e recortando a base 
df = pd.read_csv("previsao_de_renda.csv")
df = df[['renda', 'posse_de_veiculo', 'posse_de_imovel']]
````

````python
df.corr()
````

````python
sns.pointplot(x="posse_de_veiculo", 
              y="renda",
              data=df)
````

````python
sns.pointplot(x="posse_de_imovel", 
              y="renda",
              data=df)
````

**CONCLUSÃO**:

Sim, ambas parecem descriminar a renda de forma estatisticamente significante. Pois em ambos os gráficos a renda de quem tem carro (ou imóvel) é maior do que que a de quem não tem e em ambos os gráficos os intervalos de confiança calculados a partir dos erros padrão não se cruzam.
